```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('Ss.csv')
data2 = pd.read_csv('Ts.csv')
data = pd.concat([data1, data2])
```

```python
newout = pd.DataFrame()
for choice, model in data.groupby(['Data', 'Scoretype', 'Method', 'NWEEKS', 'fsval', 'r']):
    if model['TP'].hasnans:
        continue
    tmp = pd.DataFrame({'Data': [choice[0]], 'Scoretype': [choice[1]], 'Method': [choice[2]], 
                        'NWEEKS': [choice[3]], 'fsval':[choice[4]], 'r': [choice[5]], 'TP': [model['TP'].mean()], 
                        'BFP': [model['BFP'].mean()], 'TP1': [model['TP1'].mean()], 
                        'TP2': [model['TP2'].mean()], 'BFP1': [model['BFP1'].mean()], 
                        'BFP2': [model['BFP2'].mean()]})
    newout = pd.concat([newout, tmp])
```

```python
newout.index = range(len(newout))
newout.sort_values('TP', ascending = False).head(50)
```
